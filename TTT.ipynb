```python
# Escribimos un saludo de bienvenida
print("*************************************** \n     ¡Bienvenidas al Tic Tac Toe! \n*************************************** \nEstas son las posiciones que puedes jugar.")
# Creamos el tablero dentro de una clase, para después poder llamarlo más abajo
# Mostrar el tablero inicial con números para orientación de los jugadores
class Tablero:
    def __init__(self):
# El tablero nos nuestra los números como texto, pero no tiene una variable
        self.tablero = [str(i + 1) for i in range(9)] 

    def mostrar(self):
        ancho_total = 40
        print("\n")
        print(f"{self.tablero[0]} | {self.tablero[1]} | {self.tablero[2]}".center(ancho_total))
        print("--+---+--".center(ancho_total))
        print(f"{self.tablero[3]} | {self.tablero[4]} | {self.tablero[5]}".center(ancho_total))
        print("--+---+--".center(ancho_total))
        print(f"{self.tablero[6]} | {self.tablero[7]} | {self.tablero[8]}".center(ancho_total))
        print("\n")

# Creamos una instancia del tablero y lo mostramos:
mi_tablero = Tablero()
mi_tablero.mostrar()

# Le pedimos a las jugadoras que ingresen su nombre:
nombre_jugadora_1 = input("Por favor Jugadora 1, ingresa tu nombre: ")
nombre_jugadora_2 = input("Por favor Jugadora 2, ingresa tu nombre: ")

# Elegimos con qué ficha empieza cada jugadora:
ficha_1 = "X" 
ficha_2 = "O"
# Definimos variables para más adelante poder intercalar los turnos:
turno_actual = nombre_jugadora_1
# Definimos que la primera jugadora siempre arranque con la ficha "X"
# Y definimos la variable para poder intercalar más adelante:
ficha_actual = ficha_1

# Escribimos las condiciones para determinar ganadora metiéndolas en un loop:
while True:
# Mostrar de quién es el turno y con qué ficha juega
    print(f"Es el turno de {turno_actual}, tu ficha es {ficha_actual}.")

# Pedirle que haga su jugada
    posicion = input("Elige el número donde quieres jugar tu ficha:")
# Nos aseguramos de que lo que ingresa la jugadora sea un número
    if not posicion.isdigit():
        print("Por favor, ingresa un número válido.")
        continue
# Nos aseguramos de que ingresa un número entre el 1 y el 9 (incluídos)
    posicion = int(posicion) - 1
    if posicion < 0 or posicion > 8:
        print("Posición fuera de rango. Intenta de nuevo.")
        continue
# Si un espacio del tablero está ocupado por una de las dos fichas, entonces se le lanza el mensaje:
    if mi_tablero.tablero[posicion] in [ficha_1, ficha_2]:
        print("Esa posición ya está ocupada. Intenta de nuevo.")
        continue

# Actualizar el tablero con la ficha actual (lo carga pero no lo muestra):
    mi_tablero.tablero[posicion] = ficha_actual

# Mostrar el tablero actualizado (lo mostramos):
    mi_tablero.mostrar()

# Verificar condiciones de victoria:
    if (
        (mi_tablero.tablero[0] == mi_tablero.tablero[1] == mi_tablero.tablero[2] != ' ') or
        (mi_tablero.tablero[3] == mi_tablero.tablero[4] == mi_tablero.tablero[5] != ' ') or
        (mi_tablero.tablero[6] == mi_tablero.tablero[7] == mi_tablero.tablero[8] != ' ') or
        (mi_tablero.tablero[0] == mi_tablero.tablero[3] == mi_tablero.tablero[6] != ' ') or
        (mi_tablero.tablero[1] == mi_tablero.tablero[4] == mi_tablero.tablero[7] != ' ') or
        (mi_tablero.tablero[2] == mi_tablero.tablero[5] == mi_tablero.tablero[8] != ' ') or
        (mi_tablero.tablero[0] == mi_tablero.tablero[4] == mi_tablero.tablero[8] != ' ') or
        (mi_tablero.tablero[6] == mi_tablero.tablero[4] == mi_tablero.tablero[2] != ' ')
    ):
        print(f"¡Ha ganado {turno_actual}!")
        break

    # Verificar empate:
    if all(casilleros in [ficha_1, ficha_2] for casilleros in mi_tablero.tablero):
        print("¡Es un empate!")
        break

    # Cambiar de turno:
    if turno_actual == nombre_jugadora_1:
        turno_actual = nombre_jugadora_2
        ficha_actual = ficha_2
    else:
        turno_actual = nombre_jugadora_1
        ficha_actual = ficha_1
```
