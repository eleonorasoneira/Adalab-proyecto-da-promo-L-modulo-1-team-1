```python
# Crear un tablero representado por los números del 1 al 9
tablero = [str(i + 1) for i in range(9)]

# Mostrar el tablero inicial con números para orientación
print("*************************************** \n     ¡Bienvenidas al Tic Tac Toe! \n*************************************** \nEstas son las posiciones que puedes jugar.")
ancho_total = 40  # Ajusta el ancho según la consola
print("\n")
print(f"{tablero[0]} | {tablero[1]} | {tablero[2]}".center(ancho_total))
print("--+---+--".center(ancho_total))
print(f"{tablero[3]} | {tablero[4]} | {tablero[5]}".center(ancho_total))
print("--+---+--".center(ancho_total))
print(f"{tablero[6]} | {tablero[7]} | {tablero[8]}".center(ancho_total))
print("\n")


# Elegimos el nombre de cada jugadora
nombre_jugadora_1 = input("Por favor, Jugadora 1, ingresa tu nombre: ")
nombre_jugadora_2 = input("Por favor, Jugadora 2, ingresa tu nombre: ")

# Elegimos con qué ficha empieza cada jugadora
ficha_1 = "X" 
ficha_2 = "O"
turno_actual = nombre_jugadora_1
ficha_actual = ficha_1

# Condiciones para determinar ganadora
while True:
    # Mostrar de quién es el turno
    print(f"Es el turno de {turno_actual}, tu ficha es {ficha_actual}.")

    # Pedirle que haga su jugada
    posicion = input("Elige una posición del 1 al 9, siendo 5 la posición central: ")
    
    if not posicion.isdigit():
        print("Por favor, ingresa un número válido.")
        continue

    posicion = int(posicion) - 1
    if posicion < 0 or posicion > 8:
        print("Posición fuera de rango. Intenta de nuevo.")
        continue

    if tablero[posicion] in [ficha_1, ficha_2]:
        print("Esa posición ya está ocupada. Intenta de nuevo.")
        continue

    # Actualizar el tablero con la ficha actual
    tablero[posicion] = ficha_actual

    # Mostrar el tablero actualizado
    ancho_total = 40
    print("\n")
    print(f"{tablero[0]} | {tablero[1]} | {tablero[2]}".center(ancho_total))
    print("--+---+--".center(ancho_total))
    print(f"{tablero[3]} | {tablero[4]} | {tablero[5]}".center(ancho_total))
    print("--+---+--".center(ancho_total))
    print(f"{tablero[6]} | {tablero[7]} | {tablero[8]}".center(ancho_total))
    print("\n")

    # Verificar condiciones de victoria
    if (
        (tablero[0] == tablero[1] == tablero[2] != ' ') or
        (tablero[3] == tablero[4] == tablero[5] != ' ') or
        (tablero[6] == tablero[7] == tablero[8] != ' ') or
        (tablero[0] == tablero[3] == tablero[6] != ' ') or
        (tablero[1] == tablero[4] == tablero[7] != ' ') or
        (tablero[2] == tablero[5] == tablero[8] != ' ') or
        (tablero[0] == tablero[4] == tablero[8] != ' ') or
        (tablero[6] == tablero[4] == tablero[2] != ' ')
    ):
        print(f"¡Ha ganado {turno_actual}!")
        break

    # Verificar empate
    if all(casilleros in [ficha_1, ficha_2] for casilleros in tablero):
        print("¡Es un empate!")
        break

    # Cambiar de turno
    if turno_actual == nombre_jugadora_1:
        turno_actual = nombre_jugadora_2
        ficha_actual = ficha_2
    else:
        turno_actual = nombre_jugadora_1
        ficha_actual = ficha_1
```
